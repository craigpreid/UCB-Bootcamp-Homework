```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as mpatches

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
dataset = pd.merge(city_data, ride_data, how='outer', on='city')
dataset = dataset.sort_values('city')
dataset_alpha = dataset.set_index('city')

#Group by city
by_city = dataset.groupby(['city', 'type'])

# Display the data table for previous
by_city.head()
```

## Bubble Plot of Ride Sharing Data

```python
# This creates a temporary df and pulls out the values from 'type' column
# This looks weird, but I had to use type as an index to get a value I could 
# use in keys.  The city_class variable does not seem to work if I do something
# easier like use unique().  The next version has this way of storing type
transition_df = by_city.count()
city_class = transition_df.index.get_level_values('type')

# Instead of hardcoding labels, pull out unique values
labels = city_class.unique().tolist()

# This creates a key to be used to match color
key = {labels[0]: ('coral', labels[0]), labels[1]: ('lightblue', labels[1]), labels[2]:('gold', labels[2])}

```

```python
# Obtain the x and y coordinates for each of the three city types

x_max = int(round(by_city['ride_id'].count().max(),0))
x_axis = by_city['ride_id'].count()
x_lim = x_max + 2


y_max=int(round(by_city['fare'].mean().max(),0))
y_axis = by_city['fare'].mean()
y_lim = y_max + 2

size = by_city['driver_count'].count()*12

# Build the scatter plots for each city types
# This uses a for loop and values from 'key' above
plt.scatter(x_axis, y_axis, c=[key[index][0] for index in city_class], s=size, edgecolors = "black", alpha=0.60)


plt.xlim(0,x_lim)
plt.ylim(18,y_lim)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
patches = [mpatches.Circle(20,radius=5, color=color, label=label) for color, label in key.values()]
plt.legend(handles=patches, labels=[label for _, label in key.values()], loc = 'upper right')

# Incorporate a text label regarding circle sizepl
plt.text(42,38, "Note:\n Circle size correlates with driver count and city")

# Save Figure
plt.savefig("images/pyber_ride_sharing_data.png")

plt.show()
```
