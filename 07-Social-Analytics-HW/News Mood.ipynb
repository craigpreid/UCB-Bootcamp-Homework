```python
#import dependencies
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
from config import consumer_key, consumer_secret, access_token, access_token_secret
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from datetime import datetime

plt.style.use('ggplot')

analyzer = SentimentIntensityAnalyzer()
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
newsmedia = ["BBC", "CBS", "CNN", "FoxNews", "NYTimes"] #list of twitter handles used for retrieving API data

result = []

for target_user in newsmedia: #iterate through each of the new sources
    public_tweets = api.user_timeline(target_user, count = 100)

    counter = 0
    
    for tweet in public_tweets: #process each of the 100 tweets, get relevant information and append as a dictionary into a list
        raw_time = tweet["created_at"]
        converted_time = datetime.strptime(raw_time, "%a %b %d %H:%M:%S %z %Y")
        tweet_split = tweet["text"].split("https") #split the tweet text to exclude the url
        message = tweet_split[0]
        sentiments = analyzer.polarity_scores(message)
        counter += 1
        result.append({"source": target_user, "text": message, "date": converted_time, 
                       "compound": sentiments["compound"], "positive": sentiments["pos"], 
                       "neutral": sentiments["neu"], "negative": sentiments["neg"],  
                       "tweets ago": counter})

    result_df = pd.DataFrame(result)
    result_df = result_df[["source","text","date","compound","positive","neutral","negative","tweets ago"]]   

result_df.head()
```

```python
#copy the dataframe and use regular expression to strip the tweet texts of emojis and then save to csv
no_emoji =result_df.copy()

import re

def clean(mystring):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    out = emoji_pattern.sub(r'', mystring)
    return out

no_emoji['text'] = no_emoji['text'].apply(clean)

now = datetime.now()
date = now.strftime("%m-%d-%Y %H:%M").split()[0]
no_emoji.to_csv(f"Ouput/Media Tweets and Sentiment Analysis ({date}).csv", index=False, header=True)
```

```python
color_list = ["lightblue", "green", "red", "mediumslateblue", "yellow"] #list of colors for plotting

fig = plt.figure(figsize=(8,5))
ax  = fig.add_subplot(111)

#plot each of the news media's tweet sentiment against number of tweets ago
for index in range(len(newsmedia)):
    singlesource = result_df.loc[(result_df["source"] == newsmedia[index]), :]

    x_values = singlesource["tweets ago"]
    y_values = singlesource["compound"]
    color = color_list[index]
    sourcename = newsmedia[index]

    ax.scatter(x_values, y_values, facecolors=color, edgecolors="black", label = sourcename, s = 100, alpha = 0.9)

#additional plot formatting
plt.title(f"Sentiment Analysis of Media Tweets ({date})")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
    
ax.set_position([0.1,0.1,0.68,0.82])
ax.legend(loc = 'center left', bbox_to_anchor = (1.0, 0.5), title="Media Sources")
plt.xlim([x_values.max()+2,x_values.min()-2])
plt.ylim([-1,1])    
plt.savefig("Images/Sentiment Analysis of Media Tweets Scatter Plot.png")
```

```python
# Get the mean sentiment of each news source
mean_compound = result_df.groupby(["source"]).mean()["compound"]

# Create a bar plot
fig, ax = plt.subplots()
x = range(len(mean_compound))
handle = plt.bar(x, mean_compound, width = 1,align = "center")

# Additional formatting
plt.xticks(x, newsmedia) 
plt.title(f"Overall Media Sentiment based on Twitter ({date})")
plt.ylabel("Tweet Polarity")
plt.grid()

# Coloring the bars & add texts
color_index = 0
for bar in handle:
    bar.set_facecolor(color_list[color_index])
    bar.set_edgecolor("black")
    color_index += 1

plt.savefig("Images/Overall Media Sentiment based on Twitter.png")
```
